```python
lon1 = input("Enter the longitude of origin")
lat1 = input("Enter the lattitude of origin")
lon2 = input("Enter the longitude of destination")
lat2 = input("Enter the lattitude of destination")

from math import radians, cos, sin, asin, sqrt

def haversine(lat1, lon1, lat2, lon2):

      R = 6371 

      dLat = radians(float(lat2) - float(lat1))
      dLon = radians(float(lon2) - float(lon1))
      lat1 = radians(float(lat1))
      lat2 = radians(float(lat2))

      a = sin(dLat/2)**2 + cos(lat1)*cos(lat2)*sin(dLon/2)**2
      c = 2*asin(sqrt(a))

      return R * c

Answer = (float(haversine(lat1, lon1, lat2, lon2)))

print("The distance between these two points is " + str(Answer) + " kilometres.")
```

Initially, I have used the input function, allowing the user to designate values for the latitude and longitude at both the origin and destination points to provide a framework such that we can calculate the distance between said points. These points are to be labelled 'lon1','lat1','lon2','lat2'.Additioanlly we will import various mathematical methods; (radians, cos, sin, asin & sqrt) as to implement in the haversine formula, of which is used to calculate the the distance between the aforementioned co-ordinates. 
Ive identified 'R' of which denotes the Radius as 6371(km)...
